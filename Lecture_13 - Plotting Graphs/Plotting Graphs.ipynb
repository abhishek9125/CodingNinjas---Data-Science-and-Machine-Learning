```python
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0,10,2)
y = x**3
plt.plot(x,y,'bo')
plt.show()
```

```python
plt.plot(x,y,color = 'red',marker = '+',linewidth=5,label = 'X^3')
y2 = x**2
plt.plot(x,y2,color = 'black',marker = '+',linewidth=5,label = 'X^2')
plt.legend()
#plt.axis([1,100,0,200])
plt.ylabel('X^3')
plt.grid()
plt.xlabel('X')
plt.text(4,40,'Text',fontsize=12)
plt.title('Plotting Graphs')
plt.show()
```

```python

year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s = population,c = 'r',alpha = 0.5)
plt.title('Bubble Chart')
plt.show()



```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s = population,c = c,alpha = 0.5,edgecolor = 'red',marker = 'h')
plt.title('Bubble Chart')
plt.text(year[0]+0.1,salary[0] + 0.1,population[0])
plt.text(year[1]+0.1,salary[1] + 0.1,population[1])
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ['A','B','C','D']
colors = ['blue','purple','red','pink']
explode = [0.1,0.2,0,0]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct='%0.2f%%',counterclock=False,startangle=90)
plt.title('Split among Classess')
plt.axis('equal')
plt.show()
```
