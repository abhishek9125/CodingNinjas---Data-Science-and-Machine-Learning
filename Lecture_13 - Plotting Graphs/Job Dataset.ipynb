```python
# Given File 'amazonjobsdataset.csv'
# It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. 
# This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.
```

```python
# Problem Statement :
# Plot the line graph between no. of Job postings with respect to year.
# Print the year and the number of job posting as integer value.
# Note: Year should be in ascending order.
```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
import collections
with open('amazon_jobs_dataset.csv',encoding = 'UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    l = list(file_data)
    
    job = {}
    
    for row in l:
        x = row['Posting_date'].split()
        key = x[2]
        if key in job:
            job[key] += 1
        else:
            job[key] = 1
    od = collections.OrderedDict(sorted(job.items()))
    plt.plot(list(od.keys()),list(od.values()))
    plt.show()
    for i in od.keys():
        print(i,end=" ")
        print(od[i])
    
```

```python
# Plot the Bar graph between Month vs Job Openings.
# Print the year and the number of job posting as integer value.
# Order of months doesn't matter.
import matplotlib.pyplot as plt
import csv
import numpy as np
import collections
with open('amazon_jobs_dataset.csv',encoding = 'UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    l = list(file_data)
    
    job = {}
    
    for row in l:
        x = row['Posting_date'].split()
        key = x[0]
        if key in job:
            job[key] += 1
        else:
            job[key] = 1
    od = collections.OrderedDict(sorted(job.items()))
    plt.bar(list(od.keys()),list(od.values()))
    plt.xticks(rotation = 40)
    plt.show()
    for i in od.keys():
        print(i,end=" ")
        print(od[i])
    
```

```python
# Plot the scatter graph between year vs No. of jobs opening related to Java.
# Print the year and number of Jobs opening in Java Profile.
# Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. 
# Print the year in ascending order.
import matplotlib.pyplot as plt
import csv
import numpy as np
import collections
with open('amazon_jobs_dataset.csv',encoding = 'utf-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    l = list(file_data)
    
    job = {}
    
    for row in l:
        x = row['BASIC QUALIFICATIONS']
        if('Java' in x or 'java' in x):
            y = row['Posting_date'].split()
            key = y[2]
            if key in job:
                job[key] += 1
            else:
                job[key] = 1
    
    od = collections.OrderedDict(sorted(job.items()))
    plt.plot(list(od.keys()),list(od.values()))
    plt.xticks(rotation=40)
    plt.show()
    
    for i in od.keys():
        print(i,end=" ")  
        print(od[i])
    
```
