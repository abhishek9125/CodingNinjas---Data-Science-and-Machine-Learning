```python
import numpy as np
import csv

with open('terrorismData.csv',encoding='UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    l = list(file_data)
    
    state = []
    
    for row in l:
        state.append(row['State'])
    
    s = np.array(state)
    
    si = np.where(s=='Jammu and Kashmir')
    count = 0
    for i in si:
        for j in i:
            print(s[j])
```
